```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
from scipy import stats
from matplotlib.backends.backend_pdf import PdfPages
```

```python
christa=pd.read_excel('Arctic_Shrub_Christa.xlsx','Raw Data')
```

```python
christa
```

```python
props=dict(boxstyle='round',facecolor='wheat',alpha=0.5)
pp = PdfPages('christa.pdf')
fig,ax=plt.subplots()
rsq={}
ind_var='LAI'
df=christa
for dep_var in df:
    if df[dep_var].dtype==float and dep_var!=ind_var:
        ax.scatter(df[ind_var],df[dep_var])
        ax.set_xlabel(ind_var)
        ax.set_ylabel(axis_names[dep_var])
        ax.set_ylim([df[dep_var].min(),df[dep_var].max()])
        xy=df[[ind_var,dep_var]].dropna()
        x=np.array([xy.iloc[:,0].min(),xy.iloc[:,0].max()])        
        results=stats.linregress(xy.iloc[:,0],xy.iloc[:,1])
        ax.plot(x,x*results[0]+results[1])       
        textstr='m={:.3f}\nb={:.3f}\nr2={:.3f}\np={:.3f}'.format(results[0],results[1],results[2]**2,results[3])
        ax.text(0.05,0.95,textstr,transform=ax.transAxes,fontsize=10,verticalalignment='top',bbox=props)
        print dep_var,results[2]
        rsq[dep_var]=results[2]**2
        pp.savefig()
        ax.cla()

pp.close()
r2=pd.DataFrame(rsq.items(),columns=['name','rsq'])
r2.sort(columns='rsq',ascending=0,inplace=True)
#elems=r2.name[0:6]

print r2
```

```python
elems=r2.name[0:4]
ind_var='LAI'
df=christa

f, axTemp = plt.subplots(2,2)
props=dict(boxstyle='round',facecolor='white',alpha=0.5)
f.set_size_inches(10,10)
f.subplots_adjust(wspace=0.2)

for i,elem in enumerate(elems):
#for i,j in enumerate(ax):
    ax=axTemp[i%2][i/2]
    print i%2,i/2
    ax.scatter(df[ind_var],df[elem])
    ax.set_xlabel(ind_var)
    ax.set_ylabel(axis_names[elem])
    ax.set_xlim([-.1,df[ind_var].max()*1.1])
    
    xy=df[[ind_var,elem]].dropna()
    x=np.array([xy.iloc[:,0].min(),xy.iloc[:,0].max()])        
    results=stats.linregress(xy.iloc[:,0],xy.iloc[:,1])
    ax.plot(x,x*results[0]+results[1]) 
    
    textstr='m={:.3f}\nb={:.3f}\n$r^2$={:.3f}\np={:.3f}'.format(results[0],results[1],results[2]**2,results[3])
    ax.text(0.05,0.95,textstr,transform=ax.transAxes,fontsize=10,verticalalignment='top',bbox=props)
    ax.text(0.93,0.95,chr(i + ord('A')),transform=ax.transAxes,fontsize=14,verticalalignment='top')

plt.savefig('christa-4.png')
```
